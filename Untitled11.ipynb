```python
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Flatten
from tensorflow.keras.utils import to_categorical
from matplotlib import pyplot as plt
import numpy as np
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
(train_data,train_target),(test_data,test_target) = mnist.load_data()
```

```python
fig, axs = plt.subplots(2, 5, figsize=(16,8))
random_numbers = np.random.randint(0, 60000, 10)
print('Target labels: ',train_target[random_numbers])
for idx, ax in enumerate(axs.ravel()):
    ax.imshow(train_data[random_numbers[idx]],cmap='gray')
plt.show()
```

```python
new_train_data = train_data/255.0
new_test_data=test_data/255.0

new_train_target = to_categorical(train_target)
new_test_target = to_categorical(test_target)
```

```python
model=Sequential()

model.add(Flatten(input_shape=(28,28)))

# Added L1/L2 regularization to prevent overfitting
model.add(Dense(256,activation='relu'))
model.add(Dense(128,input_dim=64,activation='relu'))
model.add(Dense(64,input_dim=64,activation='relu'))
model.add(Dense(10,input_dim=32,activation='softmax'))
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
# -------------------- Train Model with Improved Hyperparameters --------------------
train_info = model.fit(new_train_data, new_train_target,
                       epochs=20,  # Increased from 10 → 20
                       batch_size=64,  # Added mini-batch training
                       validation_split=0.1,
                       verbose=1)
```

```python
plt.plot(train_info.history['loss'])
plt.xlabel('epochs')
plt.ylabel('loss')
plt.title('Model loss')
plt.show()
```

```python
plt.plot(train_info.history['accuracy'])
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.title('Model accuracy')
plt.show()
```

```python
loss, acc = model.evaluate(new_test_data , new_test_target)
print(f'Loss of the Test dataset is: {loss}\nAccuracy of the test dataset is: {acc}')
```

```python
# -------------------- Confusion Matrix --------------------
#c. Visualize class-wise test dataset performance using a confusion matrix.
y_pred_classes = np.argmax(model.predict(new_test_data), axis=1)
cm = confusion_matrix(test_target, y_pred_classes)

disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=np.arange(10))
disp.plot(cmap=plt.cm.Blues, xticks_rotation=45)
plt.title("Confusion Matrix - MNIST Classification")
plt.show()
```
